# Draw a 2D

```python
def draw_simple_2d_function(math_function):
    pio.templates.default = "none"

    x = np.linspace(-10, 10, 100)

    grapth = go.Scatter(x=x, y=math_function(x), 
                        mode='lines', 
                        line_width=2, line_color='blue', 
                        name='cosine')

    axis_style=dict(showline=False, zeroline=True, 
                    showticklabels=False, ticks='', 
                    showgrid=False)

    layout = go.Layout(width=600, height=450,
                    font_size=13,
                    xaxis = axis_style,
                    yaxis = axis_style,
                    hovermode='closest',
                    legend_x= 0,
                    legend_y=0.85
                    )

    fig = go.Figure(data=[grapth], layout=layout)
    # If you want to customize the width
    # fig.update_layout(
    #     xaxis=dict(
    #         range=[0, 1.5]
    #     )
    # )
    return fig
```

```python
import numpy as np
from numpy import pi, sin, cos
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "none"

x = np.linspace(0, 0.4, 100)

cosgraph = go.Scatter(x=x, y=cos(3*x), 
                      mode='lines', 
                      line_width=2, line_color='blue', 
                      name='cosine')

singraph = go.Scatter(x=x, y=sin(2*x), 
                      mode='lines', 
                      line_width=2, line_color='red', 
                      name='sine')


axis_style=dict(showline=False, zeroline=True, 
                showticklabels=False, ticks='', 
                showgrid=False)

layout = go.Layout(width=600, height=450,
                   font_size=13,
                   xaxis = axis_style,
                   yaxis = axis_style,
                   hovermode='closest',
                   legend_x= 0,
                   legend_y=0.85
                )

fig = go.Figure(data=[cosgraph, singraph], layout=layout)
fig.update_layout(
    xaxis=dict(
        range=[0, 1.5]
    )
)
```
